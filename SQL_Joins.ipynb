1. Retrieve all customers who have placed at least one order.
- Use INNER JOIN
- SELECT DISTINCT c.*
FROM Customers c
INNER JOIN Orders o
ON c.CustomerID = o.CustomerID;

2. Retrieve all customers and their orders, including customers who have not placed any orders.
- Use LEFT JOIN
- SELECT c.CustomerID, c.CustomerName, o.OrderID, o.Amount
FROM Customers c
LEFT JOIN Orders o
ON c.CustomerID = o.CustomerID;


3. Retrieve all orders and their corresponding customers, including orders placed by unknown
customers.
- Use RIGHT JOIN
SELECT c.CustomerID, c.CustomerName, o.OrderID, o.Amount
FROM Customers c
RIGHT JOIN Orders o
ON c.CustomerID = o.CustomerID;


4. Display all customers and orders, whether matched or not.
- Use FULL OUTER JOIN
 MySQL does not directly support FULL JOIN, so we will use UNION:
 SELECT c.CustomerID, c.CustomerName, o.OrderID, o.Amount
FROM Customers c
LEFT JOIN Orders o
ON c.CustomerID = o.CustomerID

UNION

SELECT c.CustomerID, c.CustomerName, o.OrderID, o.Amount
FROM Customers c
RIGHT JOIN Orders o
ON c.CustomerID = o.CustomerID;

5. Find customers who have not placed any orders.
- Use LEFT JOIN + NULL check
SELECT c.*
FROM Customers c
LEFT JOIN Orders o
ON c.CustomerID = o.CustomerID
WHERE o.CustomerID IS NULL;

6. Retrieve customers who made payments but did not place any orders.
- LEFT JOIN Orders + INNER JOIN Payments
SELECT DISTINCT c.*
FROM Customers c
INNER JOIN Payments p
ON c.CustomerID = p.CustomerID
LEFT JOIN Orders o
ON c.CustomerID = o.CustomerID
WHERE o.CustomerID IS NULL;

7. Generate a list of all possible combinations between Customers and Orders.
- Use CROSS JOIN
SELECT c.CustomerName, o.OrderID
FROM Customers c
CROSS JOIN Orders o;

8. Show all customers along with order and payment amounts in one table.
- SELECT c.CustomerName,
       o.Amount AS OrderAmount,
       p.Amount AS PaymentAmount
FROM Customers c
LEFT JOIN Orders o
ON c.CustomerID = o.CustomerID
LEFT JOIN Payments p
ON c.CustomerID = p.CustomerID;

9. Retrieve all customers who have both placed orders and made payments.
- Use INNER JOIN on both tables
SELECT DISTINCT c.*
FROM Customers c
INNER JOIN Orders o
ON c.CustomerID = o.CustomerID
INNER JOIN Payments p
ON c.CustomerID = p.CustomerID;
